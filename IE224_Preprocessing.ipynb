```python
DIACRITICS_MAP = {
    '√°': 'a', '√†': 'a', '·∫£': 'a', '√£': 'a', '·∫°': 'a', 'ƒÉ': 'a', '·∫±': 'a', '·∫≥': 'a', '·∫µ': 'a', '·∫∑': 'a',
    '√¢': 'a', '·∫ß': 'a', '·∫•': 'a', '·∫©': 'a', '·∫´': 'a', '·∫≠': 'a',
    '√©': 'e', '√®': 'e', '·∫ª': 'e', '·∫Ω': 'e', '·∫π': 'e', '√™': 'e', '·ªÅ': 'e', '·∫ø': 'e', '·ªÉ': 'e', '·ªÖ': 'e', '·ªá': 'e',
    '√≠': 'i', '√¨': 'i', '·ªâ': 'i', 'ƒ©': 'i', '·ªã': 'i',
    '√≥': 'o', '√≤': 'o', '·ªè': 'o', '√µ': 'o', '·ªç': 'o', '√¥': 'o', '·ªì': 'o', '·ªë': 'o', '·ªï': 'o', '·ªó': 'o', '·ªô': 'o',
    '∆°': 'o', '·ªù': 'o', '·ªõ': 'o', '·ªü': 'o', '·ª°': 'o', '·ª£': 'o',
    '√∫': 'u', '√π': 'u', '·ªß': 'u', '≈©': 'u', '·ª•': 'u', '∆∞': 'u', '·ª´': 'u', '·ª©': 'u', '·ª≠': 'u', '·ªØ': 'u', '·ª±': 'u',
    '√Ω': 'y', '·ª≥': 'y', '·ª∑': 'y', '·ªπ': 'y', '·ªµ': 'y',
    'ƒë': 'd',
    '·∫°': 'a', '√™': 'e', '√≤': 'o'
}

def clean_text(text):
    text = text.lower()

    text = unicodedata.normalize('NFD', text)

    text = ''.join([DIACRITICS_MAP.get(char, char) for char in text])

    text = re.sub(r'[^\w\s\-\>\/]', '', text)

    return unicodedata.normalize('NFC', text)
```

```python
for i in data.columns:
  data[i] = data[i].apply(clean_text)
```

```python
def convert_to_number(text):
    if 'tb' in text:
        number = float(re.sub(r'\D', '', text)) * 1024
    elif 'gb' in text:
        number = float(re.sub(r'\D', '', text))
    else:
        number = float(text)

    return number
```

```python
data['Capacity'] = data['Capacity'].apply(convert_to_number)
```

```python
provinces = [
    "ha noi", "vinh phuc", "bac ninh", "quang ninh", "hai duong", "hai phong",
    "hung yen", "ha giang", "cao bang", "bac kan", "tuyen quang", "lao cai",
    "yen bai", "thai nguyen", "lang son", "bac giang", "phu tho", "dien bien",
    "lai chau", "son la", "hoa binh", "thai binh", "ha nam", "nam dinh", "ninh binh",
    "thanh hoa", "nghe an", "ha tinh", "quang binh", "quang tri", "thua thien hue",
    "da nang", "quang nam", "quang ngai", "binh dinh", "phu yen", "khanh hoa",
    "ninh thuan", "binh thuan", "kon tum", "gia lai", "dak lak",
    "dak nong", "lam dong", "binh phuoc", "tay ninh", "binh duong", "dong nai",
    "ba ria - vung tau", "tp ho chi minh", "long an", "tien giang", "ben tre",
    "tra vinh", "vinh long", "dong thap", "an giang", "kien giang", "can tho",
    "hau giang", "soc trang", "bac lieu", "ca mau"
]
```

```python
def map_address_to_province(address, provinces):
    for province in provinces:
        if province in address:
            return province
    return np.NaN
```

```python
data['Location'] = [map_address_to_province(address, provinces) for address in data['Location']]
```
